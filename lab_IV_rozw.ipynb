## 2. Tworzenie dokumentów XML

*lxml* definiuje następujące obiekty:

* `Element` -- odpowiada elementowi z dokumentu XML;

* `ElementTree` -- odpowiada pełnemu dokumentowi XML.

Utworzymy dwoma sposobami dokument zawierający katalog biblioteczny:

```xml
<katalog>
  <książka kategoria="SF" nr="1">
    <autor>Stanisław Lem</autor>
    <tytuł>Fiasko</tytuł>
    <rok>1987</rok>
  </książka>
  <książka kategoria="dziecięca" nr="2">
    <autor>A.A. Milne</autor>
    <tytuł>Kubuś Puchatek</tytuł>

  </książka>
</katalog>
```

### 2.1. Zamiana łańcucha znaków na dokument XML 

Wykorzystaj funkcję `etree.fromstring()` do utworzenia elementu XML wprost z łańcucha znaków. Otrzymasz obiekt typu `Element`. Funkcja `etree.tostring()` przetwarza element na łańcuch bajtów. Napisz funkcję `wyświetl_pięknie(element)` wyświetlającą element z wcięciami. 

````python
s = '''<katalog>
  <książka kategoria="SF" nr="1"><autor>Stanisław Lem</autor>
<tytuł>Fiasko ęźćśż</tytuł>
<rok>1987</rok>
  </książka>
  <książka kategoria="dziecięca" nr="2">
    <autor>A.A. Milne</autor>
    <tytuł>Kubuś Puchatek</tytuł>
    <rok>1926</rok>
  </książka>
</katalog>'''
````

````python
katalog = etree.fromstring(s)
katalog
````

````python
katalog[0][0].text
````

````python
etree.tostring(katalog, encoding='cp1251')
````

````python
def wyświetl_pięknie(e, enc='utf8'):
    print(etree.tostring(e, encoding=enc, pretty_print=True, xml_declaration=True).decode(enc))
````

### 2.2. Stopniowe dodawanie składowych do dokumentu

Wykorzystaj funkcje `etree.Element()`, `etree.SubElement()` do stopniowej konstrukcji dokumentu. 

```xml
<katalog>
  <książka kategoria="SF" nr="1">
    <autor>Stanisław Lem</autor>
    <tytuł>Fiasko</tytuł>
    <rok>1987</rok>
  </książka>
  <książka kategoria="dziecięca" nr="2">
    <autor>A.A. Milne</autor>
    <tytuł>Kubuś Puchatek</tytuł>
    <rok>1926</rok>
  </książka>
</katalog>
```

````python
katalog = etree.Element('katalog')
````

````python
wyświetl_pięknie(katalog)
````

````python
fiasko = etree.SubElement(katalog, 'książka', kategoria='SF', nr='1')
````

````python
wyświetl_pięknie(katalog)
````

````python
wyświetl_pięknie(fiasko)
````

````python
autor = etree.SubElement(fiasko, 'autor')
````

````python
wyświetl_pięknie(katalog)
````

````python
autor.text = 'Stanisław Lem'
````

````python
wyświetl_pięknie(katalog)
````

### 5.2. `.iter()`

### 5.3. 

Wypisz nazwy rzek w porządku alfabetycznym. Zadbaj o usunięcie repetycji.

````python
rzeki = set()

for e in hydrologia.iter('rzeka'):
    rzeki.add(e.text)

for rzeka in sorted(rzeki):
    print(rzeka, end=' ')
````
